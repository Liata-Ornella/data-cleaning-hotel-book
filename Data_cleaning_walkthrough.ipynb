```python
#Combine to Create arrival_date
hotel_bookings['arrival_date'] = pd.to_datetime(
    hotel_bookings['original_arrival_day'].astype(str) + ' ' +
    hotel_bookings['original_arrival_month'].astype(str) + ' ' +
    hotel_bookings['original_arrival_year'].astype(str),
    format='%d %B %Y',  # Use '%d %b %Y' if month is abbreviated like "Jan"
    errors='coerce'
)




```

```python
#Separate columns
#hotel_bookings['arrival_date_month'] = hotel_bookings['arrival_date'].str.split('-', expand=True)[0]
#hotel_bookings['arrival_date_year '] = hotel_bookings['arrival_date'].str.split('-', expand=True)[1]
#hotel_bookings.head()

# 1. Extract year and month from arrival_date
hotel_bookings['arrival_date_year'] = hotel_bookings['arrival_date'].dt.year
hotel_bookings['arrival_date_month'] = hotel_bookings['arrival_date'].dt.month

# 2. Move 'arrival_date_month' to 4th column
month_col = hotel_bookings.pop('arrival_date_month')
hotel_bookings.insert(3, 'arrival_date_month', month_col)

# 3. Move 'arrival_date_year' to 5th column (which becomes 4th after above)
year_col = hotel_bookings.pop('arrival_date_year')
hotel_bookings.insert(4, 'arrival_date_year', year_col)

# 4. Rename columns (only existing ones)
hotel_bookings.rename(columns={
    'arrival_date_month': 'original_arrival_month',
    'arrival_date_year': 'original_arrival_year'
}, inplace=True)


```
